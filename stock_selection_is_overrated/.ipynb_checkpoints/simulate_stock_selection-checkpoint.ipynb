```python
def getPrice(symbol, from_date, to_date):
    API_ENDPOINT = "https://api.fyers.in/data-rest/v2/history/"
    headers = {
        "Authorization": f"{FYERS_APP_ID}:{FYERS_ACCESS_TOKEN}"
    }
    params = {
        "symbol": symbol,
        "resolution": "1D",
        "date_format": "1",
        "range_from": from_date,
        "range_to": to_date,
        "cont_flag": "1"
    }
    r1 = requests.get(API_ENDPOINT, headers=headers, params=params)
    
    data = r1.json()
    if data['s'] != 'ok':
        return pd.DataFrame()
    candles = data['candles']
    if len(candles) == 0:
        print("ErrorCode", data['code'], data['message'], symbol, from_date, to_date)
        return pd.DataFrame()
    
    df = pd.DataFrame(candles, columns=["time", "open", "high", "low", "close", "volume"])
    df['time'] = pd.to_datetime(df['time'], unit='s').dt.tz_localize(pytz.utc).dt.tz_convert('Asia/Kolkata')
    df['time'] = df['time'].dt.strftime('%Y-%m-%d %H:%M:%S')
    return df
```

```python
def getNifty100Stocks(file_path):
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Extract the values from the "Symbol" column and convert them to a list
        symbols = df['Symbol'].tolist()
        
        return symbols
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Usage example
file_path = 'ind_nifty100list.csv' #'https://nsearchives.nseindia.com/content/indices/ind_nifty100list.csv'
symbol_list = getNifty100Stocks(file_path)

# Print the list of symbols
print(symbol_list)
```

```python
def didStockUndergoSplit(df, price_change_threshold=50):
    # Calculate daily percentage change in closing price
    df['priceChange'] = df['close'].pct_change() * 100  # Percentage change

    # Find rows where the percentage change exceeds the threshold
    split_candidates = df[df['priceChange'].abs() > price_change_threshold]
    if split_candidates.shape[0] > 0:
        return True
    else:
        return False
```

```python
def simulation():
    returns = defaultdict()
    while len(returns) < max_number_of_stocks:
        symbol = random.choice(symbol_list)
        print("Randomly selected symbol:", symbol)
        # Generate a random year between 2015 and 2022
        year = random.randint(2015, 2022)
        # Generate a random day of the year (1 to 365)
        day_of_year = random.randint(1, 365)

        # Create the from_date by combining the year and day_of_year
        from_date = datetime(year, 1, 1) + timedelta(days=day_of_year - 1)

        # Calculate to_date by adding 360 days to from_date
        to_date = from_date + timedelta(days=360)

        # Format from_date and to_date as YYYY-MM-DD
        from_date_str = from_date.strftime("%Y-%m-%d")
        to_date_str = to_date.strftime("%Y-%m-%d")

        print("Random From Date:", from_date_str)
        print("Calculated To Date:", to_date_str)
        dat = getPrice('NSE:' + symbol + '-EQ', from_date_str, to_date_str)
        if dat.shape[0] == 0:
            continue
        if didStockUndergoSplit(dat):
            continue
        no_of_shares = int(investment_per_stock / dat.iloc[0]['close'])
        returns[symbol] = {
            'entry_date': dat.iloc[0]['time'][0:10], 
            'exit_date': dat.iloc[-1]['time'][0:10],
            'entry_price': dat.iloc[0]['close'],
            'exit_price': dat.iloc[-1]['close'],
            'no_of_shares': no_of_shares,
            'return': no_of_shares * (dat.iloc[-1]['close'] - dat.iloc[0]['close'])
            }
        time.sleep(1)
    return returns


```

```python
final_results = defaultdict()
for i in range(200):
    print(f'>>>>>>>>>>>>>>>>>>  {i}  <<<<<<<<<<<<<<<')
    returns = simulation()
    stock_set = '_'.join(list(returns.keys()))
    total_return = sum(stock['return'] for stock in returns.values())
    total_roi = total_return / (investment_per_stock * max_number_of_stocks) * 100
    final_results[stock_set] = total_roi
    #save the simulation to disk
    # Convert the dictionary to a DataFrame
    df = pd.DataFrame(returns).T
    # Reset the index if needed
    df.reset_index(inplace=True)
    # Rename the 'index' column to 'stock'
    df.rename(columns={'index': 'stock'}, inplace=True)
    df.to_csv(f'simulation/{stock_set}.csv', index=False)
    
```

```python
# Extract the values from the dictionary
values = list(final_results.values())

# Create a distribution plot using seaborn
sns.histplot(values, kde=True)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Distribution Plot of Values')

# Calculate mean and standard deviation
mean = sum(values) / len(values)
std_dev = (sum((x - mean) ** 2 for x in values) / len(values)) ** 0.5

# Add vertical lines for 1 and 2 standard deviations
plt.axvline(mean - std_dev, color='r', linestyle='dashed', linewidth=2, label='1 Std Dev')
plt.axvline(mean + std_dev, color='r', linestyle='dashed', linewidth=2, label='1 Std Dev')
plt.axvline(mean - 2 * std_dev, color='b', linestyle='dashed', linewidth=2, label='2 Std Dev')
plt.axvline(mean + 2 * std_dev, color='b', linestyle='dashed', linewidth=2, label='2 Std Dev')

# Show the legend
plt.legend()

plt.show()
```
