```python
def getPrice(symbol, from_date, to_date):
    API_ENDPOINT = "https://api.fyers.in/data-rest/v2/history/"
    headers = {
        "Authorization": f"{FYERS_APP_ID}:{FYERS_ACCESS_TOKEN}"
    }
    params = {
        "symbol": symbol,
        "resolution": "1D",
        "date_format": "1",
        "range_from": from_date,
        "range_to": to_date,
        "cont_flag": "1"
    }
    r1 = requests.get(API_ENDPOINT, headers=headers, params=params)
    
    data = r1.json()
    if data['s'] != 'ok':
        return pd.DataFrame()
    candles = data['candles']
    if len(candles) == 0:
        print("ErrorCode", data['code'], data['message'], symbol, from_date, to_date)
        return pd.DataFrame()
    
    df = pd.DataFrame(candles, columns=["time", "open", "high", "low", "close", "volume"])
    df['time'] = pd.to_datetime(df['time'], unit='s').dt.tz_localize(pytz.utc).dt.tz_convert('Asia/Kolkata')
    df['time'] = df['time'].dt.strftime('%Y-%m-%d %H:%M:%S')
    return df
```

```python
def getNifty100Stocks(file_path):
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Extract the values from the "Symbol" column and convert them to a list
        symbols = df['Symbol'].tolist()
        
        return symbols
    except Exception as e:
        print(f"An error occurred: {e}")
        return []

# Usage example
file_path = 'ind_nifty100list.csv' #'https://nsearchives.nseindia.com/content/indices/ind_nifty100list.csv'
symbol_list = getNifty100Stocks(file_path)

# Print the list of symbols
print(symbol_list)
```

```python
def didStockUndergoSplit(df, price_change_threshold=50):
    # Calculate daily percentage change in closing price
    df['priceChange'] = df['close'].pct_change() * 100  # Percentage change

    # Find rows where the percentage change exceeds the threshold
    split_candidates = df[df['priceChange'].abs() > price_change_threshold]
    if split_candidates.shape[0] > 0:
        return True
    else:
        return False
```

```python
investment_per_stock = 100000
```

```python
for _ in range(20):
    returns = defaultdict(defaultdict)
    for year in range(2015, 2023, 1):
        #print(year)
        random_symbols = random.sample(symbol_list, 100)
        #print(random_symbols)
        from_date_str = str(year) + '-01-01'
        to_date_str = str(year) + '-12-30'
        counter = 0
        cummulative_return = 0
        for symbol in random_symbols:
            if len(returns[year]) >= 30:
                break
            print(year, symbol)
            dat = getPrice('NSE:' + symbol + '-EQ', from_date_str, to_date_str)
            time.sleep(0.5)
            if dat.shape[0] == 0:
                continue
            if didStockUndergoSplit(dat):
                continue
            no_of_shares = int(investment_per_stock / dat.iloc[0]['close'])
            current_return = no_of_shares * (dat.iloc[-1]['close'] - dat.iloc[0]['close'])
            cummulative_return += current_return
            cummulative_roi = cummulative_return / (investment_per_stock * (counter + 1)) * 100
            returns[year][counter] = {
                'symbol': symbol,
                'entry_date': dat.iloc[0]['time'][0:10], 
                'exit_date': dat.iloc[-1]['time'][0:10],
                'entry_price': dat.iloc[0]['close'],
                'exit_price': dat.iloc[-1]['close'],
                'no_of_shares': no_of_shares,
                'return': current_return,
                'cummulative_return': cummulative_return,
                'cummulative_roi': cummulative_roi

            }
            counter += 1
    # Create a list of dictionaries with year and subkey as keys
    data = []
    for year, subkeys in returns.items():
        for subkey, subvalue in subkeys.items():
            data.append({'year': year, 'symbol#': subkey + 1, **subvalue})

    # Create the DataFrame
    df = pd.DataFrame(data)
    df.to_csv(f'simulation/simulation_{uuid.uuid1()}.csv') #save to disk

```

```python
# Define the directory path
directory_path = 'simulation'  # Replace with the actual directory path
# Use glob to find all CSV files in the directory
csv_files = glob.glob(os.path.join(directory_path, 'simulation*.csv'))
csv_files
```

```python
#See plot of any individual simulation
df = pd.read_csv(csv_files[0])
# Group by 'year' and 'symbol#' and calculate the mean of 'cummulative_roi'
mean_roi = df.groupby(['symbol#'])['cummulative_roi'].mean().reset_index()

# Define a custom color palette for the lines
custom_palette = sns.color_palette("hsv", n_colors=len(df['year'].unique()))

# Create a line chart using Seaborn with the custom color palette
plt.figure(figsize=(10, 6))
sns.lineplot(x='symbol#', y='cummulative_roi', hue='year', data=df, palette=custom_palette)

# Customize the plot
plt.xlabel('symbol#')
plt.ylabel('cummulative_roi')
plt.title('Cummulative ROI for Each Year')
plt.legend(title='Year')

# Add horizontal lines at Y = 10 and Y = 20
plt.axhline(y=20, color='black', linestyle='--', label='Y = 20')

# Show only integer values on the X-axis
plt.xticks(range(int(df['symbol#'].min()), int(df['symbol#'].max()) + 1))

# Show the plot
plt.show()
```

```python
# plot the cummulative across all years
plt.figure(figsize=(10, 6))
sns.lineplot(x='symbol#', y='cummulative_roi', data=mean_roi, marker='o')

# Customize the plot
plt.xlabel('symbol#')
plt.ylabel('Mean Cummulative ROI')
plt.title('Mean Cummulative ROI by Symbol# for one simulation')

# Show the plot
plt.show()
```

```python
# Initialize an empty list to store DataFrames from each CSV file
dfs = []

# Read and merge all CSV files into a single DataFrame
for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame (df_all)
df_all = pd.concat(dfs, ignore_index=True)

# Group by 'symbol#' and compute mean 'cummulative_roi'
mean_cummulative_roi = df_all.groupby('symbol#')['cummulative_roi'].mean().reset_index()

# Display the resulting DataFrame
# Create a line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(x='symbol#', y='cummulative_roi', data=mean_cummulative_roi, marker='o')

# Customize the plot
plt.xlabel('symbol#')
plt.ylabel('Mean Cummulative ROI')
plt.title('Mean Cummulative ROI by Symbol# aggregated across all simulations')

# Show the plot
plt.show()
```
